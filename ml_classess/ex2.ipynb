```python
import numpy as np

class LinearRegression:
    def __init__(self, learning_rate=0.1, max_iter=1000, shuffle=True, random_state=None):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.shuffle = shuffle
        self.random_state = random_state
        self.w = None  # includes bias as w[0]

    def fit(self, X, y):
        X = np.asarray(X)
        y = np.asarray(y)

        # Ensure shapes: X -> (N, k), y -> (N, 1)
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        if y.ndim == 1:
            y = y.reshape(-1, 1)

        N, k = X.shape

        # Initialize weights (bias + k feature weights)
        rng = np.random.default_rng(self.random_state)
        self.w = np.zeros((k + 1, 1))  # w[0] is bias

        for epoch in range(self.max_iter):
            idx = np.arange(N)
            if self.shuffle:
                rng.shuffle(idx)

            for i in idx:
                xi = X[i:i+1, :]          # shape (1, k)
                yi = y[i:i+1, :]          # shape (1, 1)

                y_hat = self.w[0:1, :] + xi @ self.w[1:, :]  # (1,1)
                error = yi - y_hat                            # (1,1)

                # Update bias
                self.w[0:1, :] = self.w[0:1, :] + self.learning_rate * error

                # Update feature weights
                self.w[1:, :] = self.w[1:, :] + self.learning_rate * (xi.T @ error)

        return self

    def predict(self, X):
        X = np.asarray(X)
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        return (self.w[0:1, :] + X @ self.w[1:, :]).ravel()
```
